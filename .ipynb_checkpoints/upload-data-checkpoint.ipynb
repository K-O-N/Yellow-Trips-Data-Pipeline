```python
# Function to insert data
def insert_df_in_chunks(df, table_name, engine, chunk_size=10000):
    """
    Inserts a pandas DataFrame into a SQL table in chunks.

    Args:
        df (pd.DataFrame): The DataFrame to insert.
        table_name (str): The name of the SQL table.
        engine (sqlalchemy.engine.Engine): The SQLAlchemy engine for the database connection.
        chunk_size (int, optional): The number of rows to insert in each chunk. Defaults to 10000.
    """
    # Insert the DataFrame in chunks
    for i in range(0, len(df), chunk_size):
        df_chunk = df[i:i + chunk_size]
        df_chunk.to_sql(
            name=table_name,
            con=engine,
            if_exists='append',  # Append data to the table
            index=False,         # Don't write the DataFrame index to the SQL table
            chunksize=chunk_size  # explicitly set chunksize in to_sql
        )
        print(f"Inserted chunk {i // chunk_size + 1} of {len(df) // chunk_size + 1}") # give a progress
```

```python
table_name = 'yellow_taxi_data'
insert_df_in_chunks(df, table_name, engine)
```
